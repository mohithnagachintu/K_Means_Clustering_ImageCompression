# 1. Finding the Closest Centroid 

```python
# finding the closest centroid for each training example 

# CENTROID ASSIGNMENT STEP  
# for each training example we will check for the closest centroid by using distance formula 

# THIS CODE IS VECTORIZED CODE  it can also be done by using for loops 


def findClosestCentroid(X,centroids):
    (m,n)=X.shape                   # store the X dimensions for further use
    (K,n)=centroids.shape           # also store the K dimensions for further use 
    
    # we have to find index vector which stores indices of centroids for each training example
    
    index = np.zeros((m,1),int)     # initialize the vector with zeros where size of vector is equal to (#train_examples,1)
    
    for i in range(1,m):            # loop over each example
        diff = X[i,:] - centroids   # find the difference between the training example and centroids
        diff = diff * diff   
        d = np.sum(diff,axis=1)     # diff is a K X n matrix and we sum each row w.r.t to each centroid i.e horizontally (axis = 1)
        k = np.argmin(d,axis=0)     # argmin returns the index of minimum value along the axis 
        index[i]= k                 # assign the index returned by argmin to index vector 
    return (index+1)                # index+1 is for our representation because we choose centroids numbered from 1 
```

```python
idx = findClosestCentroid(X,initial_centroids)
```

```python
print(idx[0:3])           
```

# 2. Compute Centroid

```python
#  compute centroids or MOVE CENTROID STEP 

# compute the mean of the points that are assigned to each cluster and move the centroid to mean 

def computeCentroid(X,index,K):
    (m,n) = X.shape
    centroids = np.zeros((K,n),float)   # we have K centroids since K clusters and each centroid has n features 
    for i in range(1,K+1):              # loop over each cluster 
        logical_array = (index == i)             # logical array whether index equals to corresponding cluster
        idx = np.where(logical_array == 1)       # np.where is useful to get the indices for condition logical_array== 1 
                                                 # idx will contain the indices where logical_array has 1 value
        idx = np.array(idx)                      # idx is a tuple (returned by np.where) , change it to numpy 
        idx = idx[0,:]                           # we only need row indices 
        idx = idx.reshape(-1,1)                  # 1D ---> 2D numpy array 
        x = X[idx[:,0],:]                        # get the values of the X based on the indices & assign to 'x'
        m = np.array(np.mean(x,axis = 0))        # compute the mean of x vertically(axis =0) 
        
        centroids[i-1,:]=m                       # assign the mean to the centroid 
    return centroids
```

```python
centroids = computeCentroid(X,idx,K)
```

```python
def plotDataPoints(X,index,K):
    x1 = X[:,0]
    x2 = X[:,1]
    plt.scatter(x1,x2, s=20, c=index[:,0],cmap='prism')
    cbar = plt.colorbar()
    cbar.set_label('indices')
```

```python
def plotKMeansProgress(X,centroids,previous_centroids,index,K,i):
    plotDataPoints(X,index,K)
    plt.scatter(previous_centroids[:,0],previous_centroids[:,1],color = 'red',marker = 'X',s = 100)
    plt.scatter(centroids[:,0],centroids[:,1],color = 'black',marker='X',s=100)
    plt.title(i)
    
```

```python
# Main K Means Function 

def runKMeans(X,initial_centroids,max_iters,plot_progess):
    (m,n) = X.shape 
    (K,n) = initial_centroids.shape
    centroids = initial_centroids
    previous_centroids = centroids
    index = np.zeros((m,1))
    for i in range(1,max_iters+1):
        index = findClosestCentroid(X,centroids)
        plt.figure(i)
        if(plot_progess):
            plotKMeansProgress(X,centroids,previous_centroids,index,K,i)
            previous_centroids = centroids

        centroids = computeCentroid(X,index,K)
    cent_index = {'centroids':centroids,'index':index}
    return cent_index
```

```python
cent_index = runKMeans(X,initial_centroids,max_iters,True)
```

```python
def random_initialise(X,K):
    (m,n) = X.shape
    r_idx = np.random.permutation(m)
    r_k = r_idx[:K]
    centroids = X[r_k,:]
    return centroids
```

```python
centroids = random_initialise(X,K)
```
